```python

import matplotlib.pyplot as plt 
import numpy as np 

# minSup
x = [0.1 ,0.15, 0.2, 0.25, 0.3] 

y_algorithm2 = [369, 130, 154, 52, 51] 
y_algorithm3 = [142, 65, 74, 28, 33] 
  
# second plot with x1 and y1 data 
plt.plot(x, y_algorithm2, '-o') 
plt.plot(x, y_algorithm3, color = "red", marker = 'x') 
plt.legend(["Algorithm 2", "Algorithm 3"], loc ="upper right")
plt.xlabel("Minimum support minSup") 
plt.ylabel("Runtime (s)") 
plt.title('connect') 
plt.grid()
plt.show() 
```
